### Preparation for cleaning, pre-processing & setting the selection benchmark

```python
#defining helper class, function for structuring data and ML models
class data_model():
    def __init__(self,data,d_type, **kwargs): 
        l1=list()
        self.scaler=None
        self.poly=kwargs.pop("poly", None)
        self.poly_func=kwargs.pop("poly_func", None)
        for key in kwargs: 
            #self.key=data[kwargs[key]]
            setattr(self,key,data[kwargs[key]])
            #print(key,kwargs[key])
            l1.append(kwargs[key])
        self.data=data.drop(columns=l1)
        self.model=None
        self.scaled=None
        self.type=d_type
    #add scaler function to the data to keep track of it 
    def add_scaler(self,scalerfunc):
        self.scaler=scalerfunc
        self.scaled=0
    #function to check the scaler function & return it.
    def get_scaler(self):
        if self.scaler == None :
            raise ValueError ("scaler function doesn't exist")
        return self.scaler
    def fit_scaler(self):
        scaler=self.get_scaler()
        scaler.fit(self.data[["Item_Weight","Item_Visibility","Item_MRP"]])
    #function to scale the values in the data to same range the data 
    def scale_model(self):
        scaler=self.get_scaler()
        if self.scaled ==1:
            return self.data
        else: 
            self.data[["Item_Weight","Item_Visibility","Item_MRP"]]= \
            scaler.transform(self.data[["Item_Weight","Item_Visibility","Item_MRP"]])
            self.scaled=1
            return self.data
    #function to provide the data for prediction
    def get_predict_data(self):
        if self.poly_func:
            data=self.poly_func.transform(self.data)
        else :
            data=self.data
        return data
    #function to provide the data for training
    def get_train_data(self,d_type):
        if d_type == "train" :
            data=self.data
            if self.poly:
                setattr(self,"poly_func",PolynomialFeatures(self.poly))
                data=self.poly_func.fit_transform(data)
            X_train, X_test, y_train, y_test, train_ID, test_ID = train_test_split(data,  
                                                               self.result, 
                                                                self.UID, 
                                                               test_size=0.2, random_state=43)
            return X_train, X_test, y_train, y_test
            
        
#function to run the model to train or predict from the data   
def run_model(data_model, m_type, mode="predict"):
    data=data_model.scale_model()
    model=get_model(data,m_type=m_type)
    if mode=="train" : 
        X_train, X_test,Y_train, Y_test= data_model.get_train_data(d_type=mode)
        if m_type=="tf": 
            model.fit(X_train.values, Y_train.values, epochs=600)
        if m_type=="tf":
            results=model.evaluate(X_train.values,Y_train.values)
            model.train_error=math.sqrt(results[1])
            print("Validation error is") 
            results=model.evaluate(X_test.values,Y_test.values, verbose=2)
            model.test_error=math.sqrt(results[1])
        else:
            model.fit(X_train,Y_train)
            model.train_error=get_error_function(Y_train,model.predict(X_train))
            model.test_error=get_error_function(Y_test,model.predict(X_test))
            print("Train error is ", get_error_function(Y_train,model.predict(X_train)))
            print("Validation error is ", get_error_function(Y_test,model.predict(X_test)))
            
    if mode=="predict": 
        return model.predict(data_model.get_predict_data())
    return model 
 
    
#function to save the model types and to be called upon runtime   
def get_model(data,m_type):
    if m_type=="lasso":
        model=Lasso(alpha=100)
        return model
    elif m_type=="linear":
        model=LinearRegression()
        return model
    elif m_type=="DTR":
        model=DecisionTreeRegressor(min_samples_leaf=5,random_state=42)
        return model
    elif m_type=="Ada":
        model=AdaBoostRegressor(random_state=42,n_estimators=1000, learning_rate=0.1, loss="square")
        return model
    elif m_type=="Ridge": 
        model=Ridge(alpha=300, random_state=21)
        return model
    elif m_type=="Ridge-P":
        #poly=PolynomialFeatures(4)
        #data=poly.fit_transform(data)
        model=Ridge(alpha=10,random_state=21)
        return model
    elif m_type=="linear-p":
        #poly=PolynomialFeatures(3)
        #data=poly.fit_transform(data)
        #print(data.shape,"during additional polynomials")
        model=LinearRegression()
        return model
    elif m_type=="tf":
        model=tf.keras.models.Sequential([
        tf.keras.layers.Dense(512,input_shape=(32,),activation= "relu", kernel_initializer='normal'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(128, activation= "relu", kernel_initializer='normal'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(64, activation= "relu", kernel_initializer='normal'),
        tf.keras.layers.Dense(1,)
        ])
        model.compile(optimizer="adam",
                      loss="mse",
                      metrics=["mean_squared_error"])
        return model
    else :
        raise ValueError("Model code doesn't exist")
    
def get_error_function(y, pred): 
    return math.sqrt(mean_squared_error(y,pred))
        
```

```python
#building basic cleaning data function
def process_clean_data (data_model,*args):
    #cleaningsteps
    #irregular terms (eg: LF) to standard respresentatitve terms (eg: Low Fat)
    data_model.data["Item_Fat_Content"][data_model.data["Item_Fat_Content"]=="LF"]="Low Fat"
    data_model.data["Item_Fat_Content"][data_model.data["Item_Fat_Content"]=="reg"]="Regular"
    data_model.data["Item_Fat_Content"][data_model.data["Item_Fat_Content"]=="low fat"]="Low Fat"
    '''#Merging Outlet desriptors into one column
    outlet_descriptors= ["Outlet_Type", "Outlet_Size","Outlet_Establishment_Year","Outlet_Location_Type"]
    temp=np.empty(data_model.data[outlet_descriptors].shape)
    for idx,item in enumerate(outlet_descriptors): 
        labels = factorize_custom(series=data_model.data[item],name=item)
        temp[:,idx]=labels
    temp=temp.astype("U1")
    column=list()
    for items in temp:
        sep=""
        sep=sep.join(items)
        column.append(sep)
    data_model.data["Outlet_Code"]=column
    data_model.data=data_model.data.drop(columns=outlet_descriptors)'''
    
    #Removing Nan entries from Item_Weight Column
    if args==():
        imp=IterativeImputer(max_iter=10,random_state=42, initial_strategy="mean")
    else:
        imp=args[0]
    data_x =weight_imputation(data_model, imp)
    data_model.imp=imp
    data_model.data=data_x
     
        

    #Storing index keys into external object 
    #keys=list(data.keys())

    # Creating Variables to utilize them during processing 
    #Categorical_index=[1,3,5,6,7,8]
    #category_integer=Categorical_index[2]
    #continuous_data=[0,2,4]
    
    #condition to handle the submission result as it doesn't have any output_sales column
    #if "Item_Outlet_Sales" in keys : 
    #    Output_index=9
    #    id_index=10
    #else : 
    #    id_index=9
    #    Output_index=None
    
    #creating one_hot_encoding for categorical data 
    data_model.data=pd.get_dummies(data_model.data, dummy_na=False)
    
    return data_model


def get_list_dict():
    list_dict={"Outlet_Type": ['Supermarket Type1', 'Supermarket Type2', 'Grocery Store','Supermarket Type3'],
           "Outlet_Size": ['Medium', 'High', 'Small', np.nan],
           "Outlet_Establishment_Year" : [1999, 2009, 1998, 1987, 1985, 2002, 2007, 1997, 2004], 
           "Outlet_Location_Type" : ['Tier 1', 'Tier 3', 'Tier 2']}
    return list_dict


#function to convert string into desired numeric for caategory aggregation:
def factorize_custom(series,name): 
    list_dict=get_list_dict()
    if name in list_dict:
        temp_list=list_dict[name]
        return series.apply(lambda x : temp_list.index(x))       
        
def weight_imputation(data_model, imp): 
    data_x=data_model.data.copy()
    item_list=["Item_Weight","Item_Fat_Content","Item_Type"]
    roll_list= ["Item_Fat_Content","Item_Type"]
    data_x[roll_list]=integerize_items(data_x[roll_list])
    #for idx,item in enumerate(data_x[item_list].drop(columns=["Item_Weight"]).columns):
    #    categ=data_x[item].astype("category").cat.categories
    #    categ_list.append(categ)
    #    categ_index=list(range(len(categ)))
    #    data_x[item]=data_x[item].astype("category").cat.rename_categories(categ_index)
    imp.fit(data_x[item_list][~data_model.data.Item_Weight.isna()])
    data_x[item_list]=imp.transform(data_x[item_list])
    data_x[roll_list]=de_integerize(data_x[roll_list])
    #for idx,item in enumerate(data_x[item_list].drop(columns=["Item_Weight"]).columns):
    #    data_x[item]=data_x[item].astype("category").cat.rename_categories(categ_list[idx])
    return data_x

def integerize_items(data):
    col_list=["Item_Fat_Content","Item_Type"]
    list_dict={"Item_Fat_Content": ['Low Fat', 'Regular'], "Item_Type":['Baking Goods', 'Breads', 'Breakfast', 'Canned', 'Dairy',
        'Frozen Foods', 'Fruits and Vegetables', 'Hard Drinks',
        'Health and Hygiene', 'Household', 'Meat', 'Others', 'Seafood',
        'Snack Foods', 'Soft Drinks', 'Starchy Foods']}
    temp=pd.DataFrame()
    for items in list_dict:
        temp[items]=data[items].transform(lambda x : list_dict[items].index(x))
    return temp

def de_integerize(data):
    col_list=["Item_Fat_Content","Item_Type"]
    list_dict={"Item_Fat_Content": ['Low Fat', 'Regular'], "Item_Type":['Baking Goods', 'Breads', 'Breakfast', 'Canned', 'Dairy',
        'Frozen Foods', 'Fruits and Vegetables', 'Hard Drinks',
        'Health and Hygiene', 'Household', 'Meat', 'Others', 'Seafood',
        'Snack Foods', 'Soft Drinks', 'Starchy Foods']}
    temp=pd.DataFrame()
    for items in list_dict:
        temp[items]=data[items].transform(lambda i : list_dict[items][int(i)] )
    return temp
```
